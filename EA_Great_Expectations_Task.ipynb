##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context()

data_source = context.data_sources.add_pandas("my_data_source")
data_asset = data_source.add_dataframe_asset(name="labels_data")
```

##6. Define and Create a Data Batch

```python
batch_definition = data_asset.add_batch_definition_whole_dataframe("whole_labels_batch")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

### Expectation 1

```python
expectation_1 = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="Ground_Truth_View"
)
```

### Validate Data Against Expectation 1

```python
result_1 = batch.validate(expectation_1)
result_1
```

### Expectation 2

```python
expectation_2 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Timestamp"
)
```

### Validate Data Against Expectation 2

```python
result_2 = batch.validate(expectation_2)
result_2
```

### Expectation 3

```python
expectation_3 = gx.expectations.ExpectColumnValuesToMatchRegex(
    column="Ground_Truth_View",
    regex=r"^C_\d{3}\.png$"
)
```

### Validate Data Against Expectation 3

```python
result_3 = batch.validate(expectation_3)
result_3
```
